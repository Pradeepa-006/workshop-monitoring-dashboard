```python
import pandas as pd

# Load the dataset
df = pd.read_excel("workshop_watch_2000_rows.xlsx")

# 1. Drop rows with missing values
df.dropna(inplace=True)

# 2. Drop duplicate rows
df.drop_duplicates(inplace=True)

# 3. Convert timestamp to datetime object
df['timestamp'] = pd.to_datetime(df['timestamp'])

# 4. Remove outliers using
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

for col in ['temperature', 'vibration', 'current']:
    df = remove_outliers(df, col)

# 5. Reset index
df.reset_index(drop=True, inplace=True)

# 6.Sort by timestamp and machine_type
df.sort_values(by=['timestamp', 'machine_type'], inplace=True)

# Preview cleaned data
print(df.head())

#saving in excel
df.to_excel("cleaned_workshop_data.xlsx", index=False)
```
